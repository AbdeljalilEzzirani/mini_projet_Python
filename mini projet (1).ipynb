```python
import pandas as pd

file_path = '/home/abdeljalil/Downloads/data-sell4all.xlsx'
data = pd.read_excel(file_path)

data.columns = data.columns.str.strip()

data['Âge'] = pd.to_numeric(data['Age'], errors='coerce')
data['Dépenses des clients'] = pd.to_numeric(data['Dépenses des clients'], errors='coerce')

age_median = data['Âge'].median()
age_mean = data['Âge'].mean()

expenses_median = data['Dépenses des clients'].median()
expenses_mean = data['Dépenses des clients'].mean()

print(f"Médiane de l'âge: {age_median}")
print(f"Moyenne de l'âge: {age_mean}")
print(f"Médiane des dépenses des clients: {expenses_median}")
print(f"Moyenne des dépenses des clients: {expenses_mean}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = '/home/abdeljalil/Downloads/data-sell4all.xlsx'
data = pd.read_excel(file_path)

data.columns = data.columns.str.strip()

data['Dépenses des clients'] = pd.to_numeric(data['Dépenses des clients'], errors='coerce')

expenses_by_country = data.groupby('Pays')['Dépenses des clients'].sum().sort_values()

plt.figure(figsize=(10, 6))
expenses_by_country.plot(kind='bar', color='skyblue')

plt.title('Dépenses des clients par pays')
plt.xlabel('Pays')
plt.ylabel('Dépenses des clients (Total)')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd

file_path = '/home/abdeljalil/Downloads/data-sell4all.xlsx'
data = pd.read_excel(file_path)

data.columns = data.columns.str.strip()

data['Dépenses des clients'] = pd.to_numeric(data['Dépenses des clients'], errors='coerce')
data['Age'] = pd.to_numeric(data['Age'], errors='coerce')

data = data[data['Dépenses des clients'] >= 10]

data_cleaned = data.drop_duplicates()

selected_columns = ['Pays', 'Age', 'Genre', 'Dépenses des clients']
data_selected = data_cleaned[selected_columns]

cleaned_file_path = '/home/abdeljalil/Downloads/cleaned_data_sell4all.csv'
data_selected.to_csv(cleaned_file_path, index=False)

print("Échantillon de données nettoyées :")
print(data_selected.head())
```
